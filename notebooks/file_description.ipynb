```
import os
from langchain_community.llms import Ollama

# User root variable

# erase content in README_CRON.md
with open('/Users/nilsjennissen/PycharmProjects/directory-assistant/README_CRON.md', 'w') as f:
    f.write("")
    f.close()

llm = Ollama(model="llama3")

prompt = """Your task is to generate a short description for a file content you are shown. 
Be as precise and concise as possible.
Return only the description.

File:"""

# Read all python files in the current directory
files = [file for file in os.listdir() if file.endswith('.py')]
filenames = [file for file in files if file != 'file_description.py']

descriptions = []
# Read file one after another and generate description
for filename in filenames[:2]:
    with open(filename, 'r') as f:
        code = f.read()
        prompt += f"\n\n{code}"
        # Generate description
        description = llm(prompt)
        # filename as key and description as value
        descriptions.append({filename: description})
        prompt = ""
        f.close()


with open('/Users/nilsjennissen/PycharmProjects/directory-assistant/README_CRON.md', 'w') as f:
    f.write(f"# File Descriptions\n\n")
    for description in descriptions:
        for key, value in description.items():
            f.write(f"## {key}\n")
            f.write(f"{value}\n\n")
    f.close()

```
